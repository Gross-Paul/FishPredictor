```python
def sigmoid(x):
    return 1/1+np.exp(-x)
```

```python
def CostFunction(X,y,theta,l):
    
    m = X.shape[0]
    a1 = X 
    a1 = np.insert(a1,0,1,1)

    print(a1.shape,theta[0].shape)
    z2 = theta[0].dot(a1)
    a2 = sigmoid(z2)
    a2 = np.insert(a2,0,1,1)
    
    z3 = theta[1].dot(a2)
    a3 = sigmoid(z3)
    hx = a3
    
    d3 = a3 - np.tile(y, (7,1)).transpose()    
    d2 = d3.dot(theta[1]) * a2 * (1-a2)
    
    d = np.array([d2,d3])
    a = np.array([a1,a2])
    
    
    delta = np.zeros((2,7,7))
    delta[0:0] = d2.transpose().dot(a1)
    delta[1:1] = d3.transpose().dot(a2)
    
    D = delta/m
    D += np.multiply(theta[:][1:][:],theta[:][1:][:])[0]
    
    grad = D
    
    return grad
```
